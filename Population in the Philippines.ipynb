```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_cities_and_municipalities_in_the_Philippines"
```

```python
response = requests.get(url)
```

```python
if response.status_code == 200:
    print("Successfully fetched the webpage.")
else:
    print("Failed to retrieve the webpage.")
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
tables = soup.find_all('table')
```

```python
tables
```

```python
table2 = tables[2] 
```

```python
table2
```

```python
data = []
for row in table2.find_all('tr'):
    cols = row.find_all(['td', 'th'])  # Get both header and data cells
    cols = [ele.text.strip() for ele in cols]  # Clean up the text
    data.append(cols)
```

```python
df_cities = pd.DataFrame(data)
```
